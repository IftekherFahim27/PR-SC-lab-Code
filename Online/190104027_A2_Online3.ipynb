```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

```

```python
indices = np.arange(X.shape[0])
np.random.shuffle(indices)
X_shuffled = X[indices]
y_shuffled = y[indices]
print(X_shuffled)
print(y_shuffled)
```

```python
# Split the data into training and testing sets
X_train = X_shuffled[:100]
y_train = y_shuffled[:100]
X_test = X_shuffled[100:]
y_test = y_shuffled[100:]
```

```python


# Take input from the user
K = int(input("Enter the number of neighbors (K): "))

# Perform k-nearest neighbors classification
classifier = KNeighborsClassifier(n_neighbors=K)
classifier.fit(X_train, y_train)

correctly_classified = 0

# Initialize dictionaries to store precision and recall for class "setosa"
precision_dict = {}
recall_dict = {}

for i in range(X_test.shape[0]):
    test_sample = X_test[i]
    true_label = y_test[i]

    # Find the K-nearest neighbors and their distances
    distances, indices = classifier.kneighbors([test_sample])

    # Predict the test sample
    predicted_class = classifier.predict([test_sample])[0]

    if predicted_class == true_label:
        correctly_classified += 1

    # Calculate precision and recall for class "setosa"
    c = "Setosa"  # Class "setosa"
    true_positive = np.sum((y_train[indices[0]] == c) & (y_train[indices[0]] == predicted_class))
    false_positive = np.sum((y_train[indices[0]] != c) & (y_train[indices[0]] == predicted_class))
    false_negative = np.sum((y_train[indices[0]] == c) & (y_train[indices[0]] != predicted_class))

    precision = true_positive / (true_positive + false_positive + 1e-10)  # Adding a small constant to avoid division by zero
    recall = true_positive / (true_positive + false_negative + 1e-10)

    precision_dict[c] = precision
    recall_dict[c] = recall

# Calculate accuracy
accuracy = (correctly_classified / X_test.shape[0]) * 100
print(f"Accuracy: {accuracy:.2f}%")
print(f"Precision: {accuracy:.2f}%")
print(f"Recall: {accuracy:.2f}%")
# Plotting
metrics = ['Precision', 'Recall']
values = [precision_dict["Setosa"], recall_dict["Setosa"]]


```
